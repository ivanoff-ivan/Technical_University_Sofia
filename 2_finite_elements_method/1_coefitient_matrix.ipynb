```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import floor, sqrt
```

```python
def fill_straight_diagonals(A, start_index, end_index, value, increment):
    for i in range(start_index, end_index - increment):
        A[i][i + increment] = value
        A[i + increment][i] = value
```

```python
n = 121   # n should be x ^ 2; x=[1,3,5,7,...]
d = int(sqrt(n))
A = np.zeros((n, n))
```

## Filling up the diagonals.

```python
conditions = [[4, 0],
             [-1, 1],
             [-1, d]]  #Value;Increment

for ele in conditions:
    fill_straight_diagonals(A, 0, len(A), ele[0], ele[1])

pd.DataFrame(A)
```

## Values of the secondary diagonals must be 0, based on the dimensions of the matrix

```python
def correct_secondary_diagonals(start_index, end_index, step):
    for i in range(start_index, end_index, step):
        A[i][i + 1] = 0
        A[i + 1][i] = 0


correct_secondary_diagonals(d - 1, len(A) - 1, d)
pd.DataFrame(A)
```

## Vector of the source.

```python
B = np.zeros(n)
middle_index = floor(len(B) / 2)
B[middle_index] = 1

pd.DataFrame(B)
```

## Solution for X

```python
x = np.linalg.solve(A, B)
pd.DataFrame(x)
```

```python
# Post-processing
"""
n must be an odd number if we want to place the source exactly at the middle;
"""
a = floor(sqrt(n))  
I = x.reshape(a, a)

pd.DataFrame(I)
```

```python
plt.figure(figsize=(10,10))
plt.gca().set_aspect("equal")
plt.pcolormesh(I)
plt.show()
```
